```
def preprocess(s):
  s=s.encode('ascii', 'ignore').decode('ascii')
  #regexp=[chinese_chars, ukranian_and_russian_chars, video_time, @Name, ]
  regexp=["u'[^\u4E00-\u9FA5]*'", "r'(?i)[А-ЯЁЇІЄҐ]'", "[0-9]*:[0-9]*", "@[a-zA-Z]*" ]
  for i in regexp:
    l=(re.findall(i,s))
    #print(l)
    for j in l:
      s=s.replace(j,'')
      #print(s)
  return s
```

```
s="Juhi 0:4 and 312:43 the @dfKl fdg  я w w w wф ffdsfkj :)"
preprocess(s)
```
